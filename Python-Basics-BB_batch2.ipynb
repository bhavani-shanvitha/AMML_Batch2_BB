## Rules for declaring variable:
#####  A variable name must start with a letter or underscore character
#####  A variable name cannot start with a number
#####  A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
##### Variable names are case-sensitive (age, Age and AGE are three different variables)

### Escape Character

```python
### To insert characters that are invalid in a string we use an escape character.
### An escape character starts with a backslash ‘\’ then the character you want to insert.
### Example :
### You will get an error if you use double quotes inside a string that is surrounded by double quotes:
intro="Hello sir,is this Vinu"s home?"
print(intro)
```

```python
intro="Hello sir,is this Vinu's home?"
print(intro)
```

```python
# Here double inverted comma is a special character we want to skip using \ escape character
intro="Hello sir,is this Vinu\"s home?"
print(intro)
```

### Format Specifier

```python
### A format specifier is used during input and output.
### It is a way to tell the compiler what type of data is in a variable during taking input using print() function.
### %s : it specifies the string
### %d : It specifies the int. value.

name = "latha"
print("hi, %s!!!" % name) # %s specify the string
```

```python
name = "latha "
Time = 8
print("%s, i will come back at %d am ." % (name, Time)) #%s specify the string and %d specify
#
```
